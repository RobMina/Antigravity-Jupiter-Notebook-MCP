```python
data = [1, 2, 3, 4, 5]
x = [v / 10.0 for v in data]
y = [2 * v + 1 for v in x]
print('preprocess done')
```

```python
x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)
num = sum((xi - x_mean) * (yi - y_mean) for xi, yi in zip(x, y))
den = sum((xi - x_mean) ** 2 for xi in x)
w = num / den
b = y_mean - w * x_mean
print('trained', w, b)
```

```python
pred = [w * xi + b for xi in x]
mae = sum(abs(pi - yi) for pi, yi in zip(pred, y)) / len(y)
print('mae', mae)
```
